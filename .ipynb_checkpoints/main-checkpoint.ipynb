```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import os
import glob
```

```python
path = os.getcwd()
xlsx_files = glob.glob(os.path.join(path, "yearly_data/*.xlsx"))

df = pd.DataFrame()
for i in range(len(xlsx_files)-1):
    name = "year" + str(1995+i) + '.xlsx'
    data = pd.read_excel("yearly_data/" + name)
    df = df.append(data, ignore_index=True)
    
df.to_excel("yearly_data/output.xlsx")
```

```python
df.plot(x = 'Date', y = 'A1')
```
